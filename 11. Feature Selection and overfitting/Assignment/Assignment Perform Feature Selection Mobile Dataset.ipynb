```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split as tts
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectKBest,f_classif,f_regression,chi2
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv('mobile data.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
x=df.drop(['wifi'],axis=1)
y=df['wifi']
```

### Before Use Algorithm at first check score on DecesionTree Classification()

```python
xtrain,xtest,ytrain,ytest=tts(x,y,test_size=.30,random_state=42)
```

```python
max_depth=[n for n in range(1,30)]
for depth in max_depth:
    clf=DecisionTreeClassifier(max_depth=depth)
    clf.fit(xtrain,ytrain)
    test_Accuracy=clf.score(xtest,ytest)
    train_Accuracy=clf.score(xtrain,ytrain)
    print('Depth =%d , Training Accuracy =%f , Testing Accuracy =%f '%(depth,train_Accuracy,test_Accuracy))
```

```python
clf=DecisionTreeClassifier(max_depth=20)
```

```python
clf.fit(xtrain,ytrain)
```

```python
clf.score(xtest,ytest)
```

# Use Select K Best 

```python
model=SelectKBest(score_func=f_classif)
```

```python
feature_score=model.fit(x,y)
```

```python
feature_score.scores_
```

```python
col2=pd.DataFrame(feature_score.scores_,columns=['Feature_scores'])
col1=pd.DataFrame(x.columns,columns=['Feature'])
gains=pd.concat([col1,col2],axis=1)
```

```python
gains
```

```python
new_x=gains.nlargest(15,'Feature_scores')
```

```python
new_x
```

```python
newDataframe=pd.DataFrame()
for i in new_x['Feature']:
    newDataframe=pd.concat([newDataframe,pd.DataFrame(x[i],columns=[i])],axis=1)
newDataframe
```

```python
xtrain1,xtest1,ytrain1,ytest1=tts(newDataframe,y,test_size=.30,random_state=42)
clf.fit(xtrain1,ytrain1)
clf.score(xtest1,ytest1)
```

# Try use PCA  (Principle Component Analysis)

```python
mmx=MinMaxScaler()
```

```python
scaled_x=mmx.fit_transform(x)
```

```python
scaled_x.shape
```

```python
scaled_x
```

```python
pca=PCA(n_components=3)
```

```python
x_pca=pca.fit_transform(scaled_x)
```

```python
x_pca
```

```python
features=pd.DataFrame(x_pca,columns=['pca1','pca2','pca3'])
features
```

```python
xtrain1,xtest1,ytrain1,ytest1=tts(features,y,test_size=.30,random_state=42)
clf.fit(xtrain1,ytrain1)
clf.score(xtest1,ytest1)
```
