```python
import matplotlib.pyplot as plt # it is a collection of functions that make matplotlib work like MATLAB
import pandas as pd
import seaborn as sns # for  data visualization
```

```python
df=pd.read_csv('global_laptop_selling_data.csv',encoding='latin-1') # for read CSV File Data
```

```python
df.head()
```

```python
df.nunique() # for generate All Unique Value On Dataset
```

```python
df.isnull().sum() # for here check how many column value empty
```

```python
# Now No Needed Laptop_ID COlumn so we can remove this column
df.drop('laptop_ID',axis=1,inplace=True)
```

```python
df.head()
```

# Preprocessing 

```python
df.Weight=df.Weight.str.replace("kg","") # Here Replace KG to Empty String Each Weight COlumn Data
df.Weight=pd.to_numeric(df.Weight) # Here convert Weight String Data TO Numeric Data using Pandas Library
```

```python
df.head()
```

```python
# We Can write anonymous function for handle replace Any Data and can tell data type like int or double

df.Ram=df.Ram.apply(lambda x:str(x).replace("GB","")).astype(int)
```

```python
df.head()
```

```python
df.info() # for Check All COlumn Info and tell sumery each column
```

```python
df_product_price=df.groupby('Company')['Price_euros'].mean() # for calculate Company Wise Price Mean()
```

```python
df_product_price # show Output company wise price Mean()
```

```python
# for Short Product Price and Create DataFrame
df_product_price=pd.DataFrame(data=df_product_price)
df_product_price.columns=['product Avg Price']
df_product_price.sort_values(by='product Avg Price',ascending=False,inplace=True)
```

```python
df_product_price
```

```python
plt.figure(figsize=(15,8)) # for set Figure Size
splot=sns.barplot(data=df_product_price,x=df_product_price.index,y='product Avg Price',palette='autumn') # for create Barchart
```

```python
plt.figure(figsize=(15,8)) # for set Figure Size
splot=sns.barplot(data=df_product_price,x=df_product_price.index,y='product Avg Price',palette='autumn') # for create Barchart
for p in splot.patches: # for set each value On Barchart
    splot.annotate(format(p.get_height(), '.2f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
```

```python
plt.figure(figsize=(15,7)) # for set Figure Size
splot=sns.barplot(data=df_product_price,x=df_product_price.index,y='product Avg Price',palette='autumn') # for create Barchart
for p in splot.patches: # for set each value On Barchart
    splot.annotate(format(p.get_height(), '.2f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

plt.xlabel("Name of the company",size=15)
plt.ylabel("Average Product Price",size=15)
plt.title("Average Price Per Company")
```
