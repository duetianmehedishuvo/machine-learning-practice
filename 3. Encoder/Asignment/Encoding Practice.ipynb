# 4. Ordinal Encoding Technique

```python
ordi=OrdinalEncoder() # for create a object for Ordinal Encoder
```

```python
# same 2 way handle it each column operation
# Method 01  

sex=df8.sex.unique() # for generate Sex COlumn Unique Value
ordi.categories=[sex] # for Categories Sex COlumn Data for Operation
encoded_values=ordi.fit_transform(df8[['sex']]) # for Data Fit and Transform
encoded=pd.DataFrame(encoded_values,columns=['Sex New Value']) # for generate a new Dataframe
df8['sex']=encoded # for replace All Sex Column Data to Sex New Data COlumn
```

```python
df8.head()
```

```python
# Method 02  using for Loop
for col in df9.columns:
    if is_numeric_dtype(df9[col]): # for check COlumn Value Number or not
        continue 
        
    ordi.categories=[df9[col].unique()]
    encoded_values=ordi.fit_transform(df9[[col]]) # for Data Fit and Transform
    encoded=pd.DataFrame(encoded_values,columns=[col+' ']) # for generate a new Dataframe
    df9[col]=encoded # for replace All Sex Column Data to Sex New Data COlumn
```
