# Split the DataSet

```python
x=df[['area']]
y=df[['rent']]
```

```python
x.head()
```

```python
x
```

```python
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=.30)
```

```python
yTrain.head()
```

```python
xTest
```

```python
yTest.head()
```

# Model Creation

```python
reg=LinearRegression() # for creating Object 
```

```python
reg.fit(xTrain,yTrain)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
m=reg.coef_
```

```python
c=reg.intercept_
```

```python
x1=2700
y=m*x1+c
y
```

```python
reg.predict([[2700]])
```

```python
df['predicted results']=reg.predict(x)
```

```python
df.head()
```

```python
df
```

```python
pred=reg.predict(xTest)
pred
```

```python
# for generate Mean Square Error
mse=mean_squared_error(yTest,pred)
mse
```

```python
mae=mean_absolute_error(yTest,pred)
mae
```

# Draw Best Fit Line
